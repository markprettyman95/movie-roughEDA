### movie budget csv (tn.movie_budgets.csv.gz)

```python
movie_budgets = pd.read_csv("zippedData/tn.movie_budgets.csv.gz")
movie_budgets
```

```python
movie_budgets.dtypes
```

```python
movie_budgets.production_budget = movie_budgets.production_budget.apply(lambda x: x.strip('$').replace(',', '')).astype(float)
```

```python
movie_budgets.domestic_gross = movie_budgets.domestic_gross.apply(lambda x: x.strip('$').replace(',', '')).astype(float)
```

```python
movie_budgets.worldwide_gross = movie_budgets.worldwide_gross.apply(lambda x: x.strip('$').replace(',', '')).astype(float)
```

The two above function's purpose is to prepare the foreign and domestic gross data as floats to perform mathematical operations on.

```python
movie_budgets.dtypes
```

```python
movie_budgets['total_gross'] = movie_budgets['domestic_gross'] + movie_budgets['worldwide_gross']
```

```python
movie_budgets['net'] = movie_budgets['total_gross'] - movie_budgets['production_budget']
movie_budgets
```

```python
top_100_net = movie_budgets.sort_values(by=['net'], ascending=False).head(100)
top_100_net
```

```python
genres_budget_studio = top_100_net.merge(bom_movie_gross, left_on='movie', right_on='title' ).drop_duplicates(subset='movie')
genres_budget_studio
```

```python
genres_budget_studio.groupby('studio')['net'].mean()
```

The above table represents the top 10 highest grossing movie studios after production costs

```python
num_votes = pd.read_sql("""
SELECT primary_title, genres, averagerating, numvotes
FROM movie_basics
JOIN movie_ratings ON movie_basics.movie_id = movie_ratings.movie_id
ORDER BY numvotes DESC
""", conn)
num_votes
```

```python
num_votes.dropna()
```

```python
num_votes['genre_list'] = num_votes['genres'].str.split(',')
num_votes
```

```python
genre_votes = num_votes.explode('genre_list')
genre_votes
```

```python
genre_votes.plot(kind='line', x='numvotes', y='averagerating', xticks=[0,5,10], figsize=(30,5))
```

```python
genre_votes_plot = genre_votes.groupby('genre_list')['numvotes'].sum().sort_values(ascending=False)
genre_votes_plot
```

```python
genre_votes_plot.plot(kind='barh', title='Number of Votes by Genre (in hundred millions)', xlabel='Genre', figsize=(15,15))
```

- I plan to use numvotes as a qualifier of movie success. If a large enough number of people took the trouble to not only consume the content, but to provide a vote as well, should be a good indicator of consumer sentiment
- In this instance, we will be analyzing the movies with the highest number of numvotes
